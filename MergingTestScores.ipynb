# Merging test scores
In this files, we are merging our main dataset with our test scores datasets.

```python
packed = pd.read_csv('data/distances_and_cds.csv')
ap_scores = pd.read_csv('data/test_scores/ap/processed.csv')
sat_scores = pd.read_csv('data/test_scores/sat/processed.csv')
```

```python
display(packed.head(1))
display(ap_scores.head(1))
display(sat_scores.head(1))
```

```python
with_ap_scores = pd.merge(packed, ap_scores, how='left', on=['school_num', 'year'])
```

```python
with_ap_scores.head()
```

```python
with_sat_and_ap_scores = pd.merge(with_ap_scores, sat_scores, how='left', on=['school_num', 'year'])
```

```python
with_sat_and_ap_scores.head()
```

Next, we will compute the yield for each high school - we will use it as a feature in one of our models.

```python
with_sat_and_ap_scores['yield'] = with_sat_and_ap_scores['enr_num'] / with_sat_and_ap_scores['adm_num']
with_sat_and_ap_scores['yield'].describe()
```

Let's save our results

```python
with_sat_and_ap_scores.to_csv('data/with_scores.csv', sep=',', index=False)
```
